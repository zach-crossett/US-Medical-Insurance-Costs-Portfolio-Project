### 3.1 Age Distribution

```python
sns.histplot(data=df, x='age', kde=True)
# Calculate the mean and median
mean_age = df['age'].mean()
median_age = df['age'].median()

# Add vertical lines for mean and median
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_age:.2f}')
plt.axvline(median_age, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median_age:.2f}')

plt.title('Age Distribution with Mean and Median')
plt.xlabel('Age')
plt.ylabel('Count')

# Show a legend with the mean and median labels
plt.legend()

plt.show()
plt.close()
```

### 3.2 Distribution of Costs per Smoker Status

```python
# Create a bar plot to differentiate charges by smoker status
sns.histplot(data=df, x='charges', hue='smoker', element='step')

plt.xlabel('Charges')
plt.ylabel('Count')
plt.title('Distribution of Charges by Smoking Status')

# Calculate the mean and median for non-smokers
mean_non_smoker = df[df['smoker'] == 'no']['charges'].mean()
median_non_smoker = df[df['smoker'] == 'no']['charges'].median()

# Calculate the mean and median for smokers
mean_smoker = df[df['smoker'] == 'yes']['charges'].mean()
median_smoker = df[df['smoker'] == 'yes']['charges'].median()

# Add vertical lines for mean and median for non-smokers
plt.axvline(mean_non_smoker, color='red', linestyle='dashed', linewidth=2, label=f'Mean Non-Smoker: {mean_non_smoker:.2f}')
plt.axvline(median_non_smoker, color='green', linestyle='dashed', linewidth=2, label=f'Median Non-Smoker: {median_non_smoker:.2f}')

# Add vertical lines for mean and median for smokers
plt.axvline(mean_smoker, color='blue', linestyle='dashed', linewidth=2, label=f'Mean Smoker: {mean_smoker:.2f}')
plt.axvline(median_smoker, color='orange', linestyle='dashed', linewidth=2, label=f'Median Smoker: {median_smoker:.2f}')

# Show the legend
plt.legend()

plt.show()
plt.close()
```

### 3.6 Categorical Variable Distributions - Sex, Smoker, and Region

```python
# Create bar plots for categorical variables: 'sex', 'smoker', 'region'
categorical_vars = ['sex', 'smoker', 'region']

for var in categorical_vars:
    sns.countplot(data=df, x=var)
    plt.title(f'{var} Distribution')
    plt.xlabel(var)
    plt.ylabel('Count')
    plt.show()
```

Next we will vizualize our quantitative variables using Boxplots and Scatterplots.

### 3.7 Boxplots - Quantitative Variables

```python
# Create boxplots for numeric variables: 'age', 'bmi', 'children', 'charges'
numeric_vars = ['age', 'bmi', 'children', 'charges']

for var in numeric_vars:
    sns.boxplot(data=df, y=var)
    plt.title(f'{var} Boxplot')
    plt.ylabel(var)
    plt.show()
```

### 3.8 Scatterplots - Quantitative Variables

```python
# Create a scatterplot matrix for numeric variables
sns.pairplot(df[numeric_vars])
plt.show()
```

### 3.9 Heatmap

Lastly, we'll want to understand the correlation between our variables.  Because 3 of our variables are categorical, we'll need to translate them into quantitative variables first.

Let's review our fields first:

```python
df.head()
```

Now using the library 'sklearn' we will translate our Categorical variables - Sex, Smoker, and Region - Quantitative variables.

```python
df['sex'] = df['sex'].apply({'male':0, 'female':1}.get)
df['smoker'] = df['smoker'].apply({'yes':1, 'no':0}.get)
df['region'] = df['region'].apply({'southwest':1, 'southeast':2, 'northwest':3, 'northeast':4}.get)
df.head()
```

Now we can plot our heatmap:

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(), annot = True)
plt.show()
```
